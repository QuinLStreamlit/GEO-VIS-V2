```python
import pandas as pd
from Functions.Excel_writer import Excel_writer
from Functions.merge_intervals import merge_intervals
```

```python
data_main = pd.read_excel("Lab_summary_final.xlsx")
data_main.head()
```

```python
data_main.describe()
```

```python
SPT = pd.read_excel("SPT.xlsx")
SPT.head()
```

```python
data_main = data_main.merge(right=SPT, how='outer', on=['Hole_ID', 'From_mbgl', 'Report', 'Chainage', 'Surface RL (m AHD)'])
```

```python
BH_Interpretation = pd.read_excel("Input/BH_Interpretation.xlsx")
BH_Interpretation
```

```python
data_main = merge_intervals(data_main, BH_Interpretation[['Hole_ID', 'From_mbgl', 'To_mbgl', 'Geology_Orgin', 'Consistency']])

condition = data_main['Geology_Orgin'].isna()
data_main.loc[condition, 'Geology_Orgin'] = data_main['Geology_Orgin']

condition = data_main['Consistency'].isna()
data_main.loc[condition, 'Consistency'] = data_main['Consistency_geo']

Excel_writer(data_main, "Lab_summary_final.xlsx", no_of_id_column=13)
```
